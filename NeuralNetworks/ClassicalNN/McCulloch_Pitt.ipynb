```python
import numpy as np
```

```python
def NN_classic_algo(M):
    '''Receive the matrix already counting for the bias'''
    w = np.zeros(len(M[0])-1)
    while True:
        m = 0
        for sample in M:
            y = sample[-1]
            x = sample[:-1]
            if y*(w@x) <= 0:
                w = w + y*x
                m = m + 1
        if m == 0:
            break
    return w

M = np.array( 
    [np.array([2,1,1]),
    np.array([1,1,1]),
    np.array([3,1,1]),
    np.array([0.98,1,-1]),
    np.array([0.5,1,-1]),
    np.array([0.75,1,-1])]
)

w = NN_classic_algo(M)
print(w)
```
